```python
import numpy as np

import matplotlib.pyplot as plt

channel_error_probs = ['0.04', '0.05', '0.06', '0.07', '0.08', '0,09', '0.1', '0.11', '0.12', '0.13']
max_iters = [10, 20, 30, 50]
pfail_data = np.array([[3.57E-02, 3.52E-02, 3.52E-02, 3.52E-02],
                        [8.44E-02, 7.29E-02, 7.29E-02, 7.29E-02],
                        [2.52E-01, 1.56E-01, 1.55E-01, 1.55E-01],
                        [6.27E-01, 3.68E-01, 3.54E-01, 3.53E-01],
                        [9.20E-01, 7.23E-01, 6.92E-01, 6.88E-01],
                        [9.94E-01, 9.53E-01, 9.38E-01, 9.35E-01],
                        [1.00E+00, 9.97E-01, 9.95E-01, 9.95E-01],
                        [1.000e+00, 1.000e+00, 1.000e+00, 1.000e+00],
                        [1.000e+00, 1.000e+00, 1.000e+00, 1.000e+00],
                        [1.000e+00, 1.000e+00, 1.000e+00, 1.000e+00]])
# pfail_data = np.random.rand(len(channel_error_probs), len(max_iters))  # Replace with your actual data

# Create a bar graph
fig, ax = plt.subplots()

bar_width = 0.15
index = np.arange(len(channel_error_probs))

for i, max_iter in enumerate(max_iters):
    ax.bar(index + i * bar_width, pfail_data[:, i], bar_width, label=f'Max Iters {max_iter}')

ax.set_xlabel('Psym')
ax.set_ylabel('Pfail')
ax.set_title('Efecto del número máximo de iteraciones sobre\n la probabilidad de error en decodificación')
ax.set_xticks(index + bar_width * (len(max_iters) - 1) / 2)
ax.set_xticklabels(channel_error_probs)
ax.legend()

plt.show()
```

```python
# Example data for BCH code
bch_pfail_data = np.array([6.22E-04, 7.98E-02, 5.92E-01, 9.58E-01, 9.99E-01])

# Create a bar graph
fig, ax = plt.subplots()

bar_width = 0.35
index = np.arange(len(channel_error_probs[1:6]))

ax.bar(index, pfail_data[1:6, 3], bar_width, label='LDPC Max Iters 50')
ax.bar(index + bar_width, bch_pfail_data, bar_width, label='BCH')

ax.set_xlabel('Psym')
ax.set_ylabel('Pfail')
ax.set_title('Tasa de error medida del código LDPC vs la estimación para código BCH')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(channel_error_probs[1:6])
ax.legend()

plt.show()
```
